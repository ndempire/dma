```python
#5.8.	Fill the missing value with constant
df['price'] =  df['price'].fillna(-999)
df.isnull().sum()
```

```python
#5.9.	Fill the missing value with mean, median and mode
#Before applying mean, median, mode check the datatype of evety cols
#Best suited: FOR INTEGER & FLOAT: MEAN & MEDIAN
#IF OBJECT TYPE USE MODE
df = pd.read_csv(r"C:\NEEL DESAI\CMPICA\BSC-IT\Semester_6\DMA\Assignments\Assignment_01\Automobile_dataa.csv")
na1 = ["NA","N/A","Not Available"]
df = pd.read_csv(r"C:\NEEL DESAI\CMPICA\BSC-IT\Semester_6\DMA\Assignments\Assignment_01\Automobile_dataa.csv",na_values=na1)
```

```python
#Applying mean
df['price']=df['price'].fillna(df['price'].mean())
```

```python
#Applying median
df['horsepower']=df['horsepower'].fillna(df['horsepower'].median())
```

```python
#Applying mode
df['engine-type']=df['engine-type'].fillna(df['engine-type']).mode()[0]
```

```python
#Applying mode
df['num-of-cylinders']=df['num-of-cylinders'].fillna(df['num-of-cylinders']).mode()[0]
```

```python
#Tip: If datatype is object if you want to convert into numeric use this
#df['price']= pd.to_numeric(df['price'], errors= 'coerce')
```

```python
#5.10.	Fill the missing value with forward fill, backward fill and interpolate
#forward fill
df.isnull().sum() #reading dataset again and checking the na
df.fillna(method='ffill', inplace=True)
```

```python
df.isnull().sum() #reading dataset again and checking the na
#backfill
df.fillna(method='bfill', inplace=True)
```
