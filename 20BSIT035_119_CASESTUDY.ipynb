```python
#datatype of all the columns of dataset
df.info()
```

```python
#as per dataset the cols are showing as object but actucal consists of numeric type so converting it into numberic
df['ssc_percentage']= pd.to_numeric(df['ssc_percentage'], errors= 'coerce')
df['hsc_percentage']= pd.to_numeric(df['hsc_percentage'], errors= 'coerce')
df['degree_percentage']= pd.to_numeric(df['degree_percentage'], errors= 'coerce')
df['emp_test_percentage']= pd.to_numeric(df['emp_test_percentage'], errors= 'coerce')
df['mba_percent']= pd.to_numeric(df['mba_percent'], errors= 'coerce')
print(df.info()) #checking whether object datatype is converted to numeric or not
```

```python
#checking for NA Values in dataframe
df.isnull().sum()
```

```python
 #printing unique and nunique values with its count
for col in df:
  print(col + " unique count = " + str(df[col].nunique()))
  print(df[col].unique(),end="\n\n")
```

```python
#changing cols name as per requirement
df=df.rename(columns={'ssc_percentage': 'ssc_percent', 'hsc_percentage': 'hsc_percent', 'mba_percent': 'pg_percent'})
```

```python
#treating all those values accordimg to their datatypes
#replace Na values of ssc percentage, Hsc percentage, Degree percentage and emp test percentage with mean
#replace Na values of ssc percentage, Hsc percentage, Degree percentage and emp test percentage with mean
df['ssc_percent']=df['ssc_percent'].fillna(df['ssc_percent'].mean())
df['hsc_percent']=df['ssc_percent'].fillna(df['hsc_percent'].mean())
df['degree_percentage']=df['degree_percentage'].fillna(df['degree_percentage'].mean())
df['emp_test_percentage']=df['emp_test_percentage'].fillna(df['emp_test_percentage'].mean())

#replacing na values of ssc_board, hsc_board, hsc subject, undergraduate degree by mode
df['ssc_board']=df['ssc_board'].fillna(df['ssc_board']).mode()[0]
df['hsc_board']=df['hsc_board'].fillna(df['hsc_board']).mode()[0]
df['hsc_subject']=df['hsc_subject'].fillna(df['hsc_subject']).mode()[0]
df['undergrad_degree']=df['undergrad_degree'].fillna(df['undergrad_degree']).mode()[0]
df['status'].fillna(method='bfill', inplace=True) #replacing by backward fill method
```

```python
#checking for NA Values in dataframe is removed or not
df.isnull().sum()
```

```python
#Getting the statistical summary
df.describe()
```

```python
#Using this to convert categorical values to numerical
df_encoded = pd.get_dummies(df, columns=['ssc_board','hsc_board','hsc_subject','undergrad_degree','work_experience','specialisation'])

# print the encoded dataframe
print(df_encoded)
```

```python
#Getting infor of encoded dataset
df_encoded.info()
```

```python
#Here X represnts the feature cols
X = df_encoded.values[:, 2:20]
#here Y represets as target cols
Y = df_encoded.values[:,1]
print(X)
print(Y)
```

```python
#IMPORTING TRAIN_TEST_SPLIT FROM sklearn
from sklearn.model_selection import train_test_split
#here we are splliting data in 70% trainging and 30%testing data

X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, random_state = 100)
```

Here there are two methods:
    1. Gini
    2. Entropy
Gini:    Gini impurity is a measure of the degree of uncertainty or impurity in a set of data, 
         and is used in decision trees to evaluate the quality of potential splits.
Entropy: Entropy is a measure of the degree of uncertainty or impurity in a set of data, 
         and is used in decision trees to evaluate the quality of potential splits.

```python
#IMPORTING decision tree classifier FROM sklearn
from sklearn.tree import DecisionTreeClassifier
#applying both gini and entropy method for checking accuracy
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,max_depth=3, min_samples_leaf=5)
clf_gini.fit(X_train, y_train)
```

```python
y_pred = clf_gini.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy is ", accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
#importing classification report from sklearn
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#IMPORTING decision tree classifier FROM sklearn
from sklearn.tree import DecisionTreeClassifier
#applying both gini and entropy method for checking accuracy
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100,
                               max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
```

```python
y_pred = clf_entropy.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
#importing classification report from sklearn
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy is ", accuracy_score(y_test,y_pred)*100)
```
