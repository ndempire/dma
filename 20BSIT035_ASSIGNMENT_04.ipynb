```python
#Displaying the count of tuples with status of missing values
na1 = ["NA","N/A","Not Available"]
df = pd.read_csv(r"C:\NEEL DESAI\CMPICA\BSC-IT\Semester_6\DMA\Assignments\Assignment_04\Wholesale_customers_data.csv",na_values=na1)
df.isnull().sum()

#as there are not any missing values values we don't need to handle missing values
```

```python
#Question 3: Perform one-hot encoding and normalization.

#Appling one hot encoding to categorical column and storing that in new df
#applying getdummies function present in basic pandas library
newdf = pd.get_dummies(df, columns = ['Channel','Region'])
print(newdf)
```

```python
#importing preprocessing from sklearn
from sklearn import preprocessing 

#Appling normalization on the newdf and storing that in nordf
nordf = preprocessing.normalize(newdf)
nordf
```

```python
#Question 4. Apply k-means clustering algorithm
#Importing kmeans for sklear
from sklearn.cluster import KMeans
kmeans = KMeans(5)
kmeans.fit(nordf)
```

```python
pred = kmeans.predict(nordf)
pred
```

```python
frame = pd.DataFrame(nordf)
frame['cluster'] = pred
frame['cluster'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder

OneHotEncoder().fit_transform(nordf)
```

```python
#We cannot do directly, for more than 3 columns. 
#Applying a Principal Component Analysis to reduce the space in 2 columns and visualize this instead.
from sklearn.decomposition import PCA

pca_num_components = 2

reduced_data = PCA(n_components=pca_num_components).fit_transform(nordf)
results = pd.DataFrame(reduced_data,columns=['pca1','pca2'])

import seaborn as sns #importing seaborn
sns.set()
import matplotlib.pyplot as plt #importing matplotlib
#plotting scatterplot
sns.scatterplot(x="pca1", y="pca2", hue=frame['cluster'], data=results)
plt.title('K-means Clustering with 2 dimensions')
plt.show()
```

```python
#5. Apply hierarchical clustering algorithm.
#lets first normalize the data and bring all the variables to the same scale:
from sklearn.preprocessing import normalize
scaleddata = normalize(df)
scaleddata = pd.DataFrame(scaleddata, columns=df.columns)
scaleddata.head()
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))   
dendogram=shc.dendrogram(shc.linkage(scaleddata, method='ward'))
plt.axhline(y=10, color='r', linestyle='--')
```

```python
#Here there are two clusters so we can apply agglomerative clustering
from sklearn.cluster import AgglomerativeClustering
#Importing agglomerative clustering from sklearn
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
#Here giving n cluser as 2
#giving affinity as auclidean and linkage as ward
cluster.fit_predict(scaleddata)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(scaleddata['Fresh'], scaleddata['Milk'], c=cluster.labels_) 
```
